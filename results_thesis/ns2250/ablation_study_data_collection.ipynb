```python
from os.path import join, dirname, exists
from os import makedirs
from thesis_v2 import dir_dict
```

```python
import pandas as pd
import numpy as np
```

```python
from thesis_v2.analysis.ablation_study import collect_all_data
```

```python
def load_original_perf():
    ret = pd.read_pickle(join(dir_dict['analyses'], 'ns2250', 'main', 'perf.pkl'))
    return ret
```

```python
def load_original_depth():
    ret = pd.read_pickle(join(dir_dict['analyses'], 'ns2250', 'main', 'source_analysis.pkl'))
    return ret
```

```python
def load_multipath_perf():
    ret = pd.read_pickle(join(dir_dict['analyses'], 'ns2250', 'multipath', 'perf.pkl'))
    return ret

def load_multipath_depth():
    ret = pd.read_pickle(join(dir_dict['analyses'], 'ns2250', 'multipath', 'source_analysis.pkl'))
    return ret
```

```python
def load_onlyDX_perf():
    ret = pd.read_pickle(join(dir_dict['analyses'], 'ns2250', 'ablation', 'perf.pkl'))
    return ret[ret.index.get_level_values('multi_path_hack').map(lambda x: x.startswith('onlyD')).values]

def load_onlyDX_depth():
    ret = pd.read_pickle(join(dir_dict['analyses'], 'ns2250', 'ablation', 'source_analysis.pkl'))
    return ret[ret.index.get_level_values('multi_path_hack').map(lambda x: x.startswith('onlyD')).values]
```

```python
def load_leDXgeDX_perf():
    ret = pd.read_pickle(join(dir_dict['analyses'], 'ns2250', 'ablation', 'perf.pkl'))
    return ret[ret.index.get_level_values('multi_path_hack').map(lambda x: x.startswith('leD') or x.startswith('geD')).values]

def load_leDXgeDX_depth():
    ret = pd.read_pickle(join(dir_dict['analyses'], 'ns2250', 'ablation', 'source_analysis.pkl'))
    return ret[ret.index.get_level_values('multi_path_hack').map(lambda x: x.startswith('leD') or x.startswith('geD')).values]
```

```python
def load_new_result():
    f_main_result = join(dir_dict['analyses'], 'ns2250', 'ablation_2L_16,32ch_1400.pkl')
    makedirs(dirname(f_main_result), exist_ok=True)
    if not exists(f_main_result):
        df_main_result = collect_all_data(
        data_loader_dict = {
            'multipath': {
                'perf': load_multipath_perf,
                'depth': load_multipath_depth,
            },
            'original': {
                'perf': load_original_perf,
                'depth': load_original_depth,
            },
            'onlyDX': {
                'perf': load_onlyDX_perf,
                'depth': load_onlyDX_depth,
            },
            'leDXgeDX': {
                'perf': load_leDXgeDX_perf,
                'depth': load_leDXgeDX_depth,
            }
        },
            train_keep=1400,
        )
        df_main_result.to_pickle(f_main_result)
        del df_main_result
    df_main_result = pd.read_pickle(f_main_result)
    return df_main_result
```

```python
df_main_result_new = load_new_result()
df_main_result_new
```

```python
def load_old_result():
    return pd.read_pickle(join(dir_dict['analyses'], 'tang_refactored', 'ablation_2L_16,32ch_1400.pkl'))
df_main_result_old = load_old_result()
df_main_result_old
```

```python
assert df_main_result_old.equals(df_main_result_new)
```
