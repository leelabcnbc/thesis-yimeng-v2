```python
from os.path import join, dirname, exists
from os import makedirs
from itertools import chain, islice
from thesis_v2 import dir_dict
from thesis_v2.configs.model.maskcnn_polished_with_rcnn_k_bl import ablation_models_ns2250_generator
from thesis_v2.analysis.io import collect_rcnn_k_bl_source_analysis
```

```python
import pandas as pd
```

```python
def load_new_result():
    f_main_result = join(dir_dict['analyses'], 'ns2250', 'ablation', 'source_analysis.pkl')
    if not exists(f_main_result):
        makedirs(dirname(f_main_result), exist_ok=True)
        df_main_result = collect_rcnn_k_bl_source_analysis(
            fixed_keys = {
            'kernel_size_l23': 3,
            'kernel_size_l1': 9,
            'ff_1st_block': True,
            'pooling_ksize': 3,
            'pooling_type': 'avg',
            'bn_after_fc': False,
            'rcnn_bl_psize': 1,
            'rcnn_bl_ptype': None,
            'input_size': 37,
            'split_seed': 'legacy',
            'dataset_prefix': 'tang',
            'model_prefix': 'maskcnn_polished_with_rcnn_k_bl',
            'scale_name': '0.01',
            'scale': '0.01',
            'smoothness_name': '0.000005',
            'smoothness': '0.000005',
            'additional_key': '0,500',
            'multi_path': True,
            'multi_path_separate_bn': True,
        },
            generator=chain(
                ablation_models_ns2250_generator(with_source=True),
            ),
            total_num_param=30,
            train_size_mapping=dict(),
            no_missing_data=False,
        )
        df_main_result.to_pickle(f_main_result)
        del df_main_result
    df_main_result = pd.read_pickle(f_main_result)
    return df_main_result
```

```python
df_main_result_new = load_new_result()
df_main_result_new
```

```python
# validate against old result
def load_old_result():
    ret = pd.concat(
        [
            pd.read_pickle(join(dir_dict['analyses'], 'tang_refactored', '20201218', 'source_analysis.pkl')),
            pd.read_pickle(join(dir_dict['analyses'], 'tang_refactored', '20201221', 'source_analysis.pkl')),
        ], axis=0
    ).sort_index()
    assert ret.index.is_unique
    return ret
```

```python
df_main_result_old = load_old_result()
```
