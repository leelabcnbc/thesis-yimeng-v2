```python
def load_new_result():
    f_main_result = join(dir_dict['analyses'], 'ns2250', 'main', 'perf_v1.pkl')
    if not exists(f_main_result):
        makedirs(dirname(f_main_result), exist_ok=True)
        df_main_result = collect_rcnn_k_bl_main_result(
            fixed_keys = {
            'kernel_size_l23': 3,
            'kernel_size_l1': 9,
            'ff_1st_block': True,
            'pooling_ksize': 3,
            'pooling_type': 'avg',
            'bn_after_fc': False,
            'rcnn_bl_psize': 1,
            'rcnn_bl_ptype': None,
            'input_size': 37,
            'split_seed': 'legacy',
            'dataset_prefix': 'tang',
            'model_prefix': 'maskcnn_polished_with_rcnn_k_bl',
            'scale_name': '0.01',
            'scale': '0.01',
            'smoothness_name': '0.000005',
            'smoothness': '0.000005',
                'additional_key': '0,500'
        },
            generator=chain(
                main_models_ns2250_generator(with_source=True),
            ),
            total_num_param=27,
            train_size_mapping=dict(),
            no_missing_data=True,
            cc_max_all_neurons=load_ccmax(),
            num_neuron=mask.sum(),
            skip_eval_json=True,
            mask=mask,
        )
        df_main_result.to_pickle(f_main_result)
        del df_main_result
    df_main_result = pd.read_pickle(f_main_result)
    return df_main_result
```
