```python
from os.path import join, dirname, exists
from os import makedirs
from itertools import chain, islice
from thesis_v2 import dir_dict
from thesis_v2.configs.model.maskcnn_polished_with_rcnn_k_bl import main_models_8k_generator
from thesis_v2.analysis.io import collect_rcnn_k_bl_main_result
from thesis_v2.analysis.imagenet8k import load_ccmax
```

```python
import pandas as pd
```

```python
def load_new_result():
    f_main_result = join(dir_dict['analyses'], 'imagenet_8k', 'main', 'perf.pkl')
    if not exists(f_main_result):
        makedirs(dirname(f_main_result), exist_ok=True)
        df_main_result = collect_rcnn_k_bl_main_result(
            fixed_keys = {
            'kernel_size_l23': 3,
            'kernel_size_l1': 9,
            'ff_1st_block': True,
            'pooling_ksize': 3,
            'pooling_type': 'avg',
            'bn_after_fc': False,
            'rcnn_bl_psize': 1,
            'rcnn_bl_ptype': None,
            'input_size': 50,
            'split_seed': 'legacy',
            'dataset_prefix': 'yuanyuan_8k_a_3day',
            'model_prefix': 'maskcnn_polished_with_rcnn_k_bl',
            'scale_name': '0.01',
            'scale': '0.01',
            'smoothness_name': '0.000005',
            'smoothness': '0.000005',
        },
            generator=chain(
                main_models_8k_generator(with_source=True),
            ),
            total_num_param=26,
            train_size_mapping={
                None: 5120,
            },
            no_missing_data=True,
            cc_max_all_neurons=load_ccmax(),
            num_neuron=79,
            skip_eval_json=True,
        )
        df_main_result.to_pickle(f_main_result)
        del df_main_result
    df_main_result = pd.read_pickle(f_main_result)
    return df_main_result
```

```python
df_main_result_new = load_new_result()
df_main_result_new
```
