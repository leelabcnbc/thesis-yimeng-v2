```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from thesis_v2.analysis.hal_analysis_refactor.model_orientation_tuning import get_stimuli_dict
```

```python
stimuli_dict = get_stimuli_dict()
```

```python
stimuli_dict2 = get_stimuli_dict(new_size=50)
```

```python
def check_stuff(idxs_this, stimuli):
    n_row, n_case = idxs_this.shape
    plt.close('all')
    fig, axes = plt.subplots(nrows=n_row, ncols=n_case, figsize=(20, 20))
    assert axes.size == n_row * n_case
    for idx_i in range(n_row):
        for idx_j in range(n_case):
            axes[idx_i][idx_j].imshow(stimuli[idxs_this[idx_i,idx_j],0], vmin=0, vmax=255, cmap='gray')
    
    plt.show()
```

```python
def plot_one_case(stimuli_dict_this):
    for name, idxs_this in stimuli_dict_this['idx_dict'].items():
        print(name)
        check_stuff(idxs_this, stimuli_dict_this['stimuli'])
```

```python
plot_one_case(stimuli_dict)
```

```python
plot_one_case(stimuli_dict2)
```

```python
from thesis_v2.analysis.hal_analysis_refactor.model_orientation_tuning import get_bars
```

```python
def plot_bars():
    bars = get_bars()
    assert len(bars) == 8
    plt.close('all')
    fig, axes = plt.subplots(nrows=1,ncols=8,figsize=(16, 2))
    axes = axes.ravel()
    assert len(bars) == len(axes) == 8
    for ax, bar in zip(axes, bars):
        ax.imshow(bar, vmin=bar.min(), vmax=bar.max(), cmap='gray')
    plt.show()
```

```python
plot_bars()
```
